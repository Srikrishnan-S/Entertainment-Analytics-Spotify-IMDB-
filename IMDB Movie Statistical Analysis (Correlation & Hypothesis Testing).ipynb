### DUPLICATE VALUES

```python
# Treating Duplicates
class ProcessDuplicates:
    def __init__(self, df):
        self.df = df

    def duplicate_count(self):
        return self.df.duplicated().sum()

    def duplicate_rate(self):
        rate = self.df.duplicated().sum() * 100 / len(self.df)
        return rate

    def detect_duplicate_features(self, features):
        if_duplicate_df = pd.DataFrame({
            "Column_Name": features,
            "If_Duplicate": [self.df[feature].duplicated().any() for feature in features]
        })
        return if_duplicate_df

    def percentage_duplicate(self, features):
        duplicate_percentage = pd.DataFrame({
            "Column Name": features,
            "Percentage Duplicate": [self.df[feature].duplicated().mean() * 100 for feature in features]
        })
        return duplicate_percentage

    def plot_duplicate(self, features, vertical=True):
        plot_df = self.percentage_duplicate(features)
        if vertical:
            plot = sns.barplot(x="Column Name", y="Percentage Duplicate", data=plot_df)
        else:
            plot = sns.barplot(y="Column Name", x="Percentage Duplicate", data=plot_df)
        return plot

    def remove_duplicates(self, features):
        for feature in features:
            self.df.drop_duplicates(subset=feature, inplace=True)
```

### NULL VALUES

```python
# Treating Nulls
class ProcessNulls:
    def __init__(self, df):
        self.df = df

    def null_count(self):
        return self.df.isnull().sum()

    def null_rate(self):
        rate = self.df.isnull().sum() * 100 / len(self.df)
        return rate

    def detect_null_features(self, features):
        if_null_df = pd.DataFrame({
            "Column_Name": features,
            "If_Null": [self.df[feature].isnull().any() for feature in features]
        })
        return if_null_df

    def percentage_null(self, features):
        null_percentage = pd.DataFrame({
            "Column Name": features,
            "Percentage Null": [self.df[feature].isnull().mean() * 100 for feature in features]
        })
        return null_percentage

    def plot_null(self, features, vertical=True):
        plot_df = self.percentage_null(features)
        if vertical:
            plot = sns.barplot(x="Column Name", y="Percentage Null", data=plot_df)
        else:
            plot = sns.barplot(y="Column Name", x="Percentage Null", data=plot_df)
        return plot

    def remove_nulls(self, features):
        for feature in features:
            self.df.dropna(subset=[feature], inplace=True)

    def impute_null(self, features, method='mean', value=None):
        for feature in features:
            if method == 'mean':
                self.df[feature].fillna(self.df[feature].mean(), inplace=True)
            elif method == 'median':
                self.df[feature].fillna(self.df[feature].median(), inplace=True)
            elif method == 'mode':
                self.df[feature].fillna(self.df[feature].mode().iloc[0], inplace=True)
            elif method == 'custom' and value is not None:
                self.df[feature].fillna(value, inplace=True)
```

**Budget by Genre :**

```python
Genre_Budget = Data.groupby('genre')['budget'].sum().sort_values(ascending = False)
Genre_Budget
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=Genre_Budget.values, y=Genre_Budget.index, palette="viridis")

plt.xlabel("Total Budget")
plt.ylabel("Genre")
plt.title("Total Budget by Genre")
plt.show()
```

**ANOVA Test :**

```python
import scipy.stats as stats

genre_budgets = [Data[Data['genre'] == genre]['budget'] for genre in Genre_Budget.index]

f_statistic, p_value = stats.f_oneway(*genre_budgets)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("There are significant differences between at least two groups.")
else:
    print("There are no significant differences between the groups.")
```
